```python
keywords = df[["Author Keywords","Index Keywords"]]
keywords.head(3)
```

```python
def split_concat(df, col_name, sep=";"):
    return pd.DataFrame(df[col_name].str.split(sep).tolist(), index=keywords.index).stack()
```

```python
keywords.dropna(inplace=True)
author_keywords=split_concat(keywords,"Author Keywords")
index_keywords=split_concat(keywords,"Index Keywords")
```

```python
keywords_df=author_keywords.append(index_keywords)
keywords_df.drop_duplicates(keep='first', inplace=True)
```

```python
keywords_df.head(3)
```

```python
keywords_df.to_csv('keywords_df.csv')
```

```python
affiliations=df[["Affiliations"]]
affiliations.head(3)
```

```python
affiliations_df=affiliations.drop('Affiliations', axis=1).join(df['Affiliations'].str.split(';', expand=True).stack().reset_index(level=1, drop=True).rename('Affiliations'))
```

```python
new = affiliations_df["Affiliations"].str.split(",", n = -1, expand = True)

affiliations_df["dept_name"]= new[0]
affiliations_df["university"]= new[1]
affiliations_df["city"]= new[2]
affiliations_df["country"]= new[3]
affiliations_df.drop(columns =["Affiliations"], inplace = True)
affiliations_df
  
```

```python
df.columns.to_list()
affiliations_df.to_csv('affiliations_df.csv')
```
